Self

self represents the instance of the class By using the "self" we can access the attributes and methods of the class in

python.

self is parameter in function and user can use another parameter name in place of it

But it is advisable to use self because it increases the readability of code

```python
class person:

  def read(self):
    print('person is reading')

  def walk(self):
    print('person is walking')

obj=person()

obj.read()
obj.walk()

obj1=person()
obj.walk()

```
